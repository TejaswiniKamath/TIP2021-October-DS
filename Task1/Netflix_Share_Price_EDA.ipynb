```python
netflix['Date']=pd.to_datetime(netflix['Date'],format='%Y-%m-%d')
netflix.info()
```

```python
fig, ax = plt.subplots(figsize=(20,6))
ax.plot(netflix['Date'], netflix['Close'], color='Green')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Stock Price', fontsize='11')
plt.title('Netflix Stock Prices')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,6))
ax.plot(netflix['Date'], netflix['Volume'], color='Green')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Volume', fontsize='11')
plt.title('Volume Trend')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,6))
high, =ax.plot(netflix['Date'], netflix['High'])
high.set_label('High')
low, =ax.plot(netflix['Date'], netflix['Low'])
low.set_label('Low')
low.set_dashes([2,2,2,2])
Open, =ax.plot(netflix['Date'], netflix['Open'])
Open.set_label('Open')
Open.set_dashes([2,2,10,2])
Close, =ax.plot(netflix['Date'], netflix['Close'])
Close.set_label('Close')
Close.set_dashes([2,2,5,2])
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Stock Price', fontsize='11')
plt.title('Netflix Stock Prices')
plt.legend()
plt.grid()
plt.show()
```

Netflix started as DVD mailing service. 

In Jan 2007, streaming services were started. 

```python
# Looking at share prices in the 6 month window before and after that
humbleBeginnings= netflix[(netflix.Date<'2010-01-01')]
fig, ax = plt.subplots(figsize=(20,6))
ax.plot(humbleBeginnings['Date'], humbleBeginnings['Close'], color='Green')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Stock Price', fontsize='11')
plt.title('Netflix Stock Prices')
plt.grid()
plt.show()
```

In July 2011, Netflix unbundled the DVD/streaming services and moved the mailing service to Quikster resulting in price rise. With users facing this sudden change, there was a massive drop in subscriptions as well as members discontinuing services. This can be seen below.

https://abcnews.go.com/Technology/netflix-earnings-stock-drops-800000-members-quit-online/story?id=14809394


```python
# Looking at share prices in the 6 month window before and after that
quiksterControversy= netflix[(netflix.Date>'2010-01-01')&(netflix.Date<'2012-01-31')]
fig, ax = plt.subplots(figsize=(20,6))
ax.plot(quiksterControversy['Date'], quiksterControversy['Close'], color='Green')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Stock Price', fontsize='11')
plt.title('Netflix Stock Prices')
plt.grid()
plt.show()
```
