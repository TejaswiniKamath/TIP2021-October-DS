```python
# function to find the % of null values in a column. This is shown in the descending order.

def findnull(df):
    totalnull = df.isnull().sum().sort_values(ascending = False)
    percentnull = (df.isnull().sum()/len(df.index)*100).sort_values(ascending = False)
    nullinfo = pd.concat([totalnull,percentnull],axis=1)
    colnames = ['Total','%']
    nullinfo.columns = colnames
    return nullinfo

# Fetching the columns null value information
findnull(lsElections)
```

Since the percentage of null values is quite less, not dropping any columns

```python
def valCounts(df,x,filt,y):
    totalCount = df[df[x]==filt][y].value_counts()
    percentCount = df[df[x]==filt][y].value_counts(normalize=True)
    countInfo = pd.concat([totalCount,percentCount],axis=1)
    colnames = ['Total','%']
    countInfo.columns = colnames
    return countInfo
```

#### Checking for top 10 highest number of candidates in a constituency 

```python
maleRep = pd.DataFrame(lsElections[lsElections['GENDER']=='MALE'].groupby(['CONSTITUENCY'])['GENDER'].count().sort_values(ascending=False).head(10).reset_index())
femaleRep = pd.DataFrame(lsElections[lsElections['GENDER']=='FEMALE'].groupby(['CONSTITUENCY'])['GENDER'].count().sort_values(ascending=False).head(10).reset_index())
print(maleRep,'\n\n' ,femaleRep)
```

```python
f,axes = plt.subplots(2,1,figsize=(24,12),sharey=True)
ax1=sns.barplot(x=maleRep.CONSTITUENCY,y=maleRep.GENDER,ax=axes[0])
plt.sca(axes[0])
plt.ylabel('Male Candidate Count')

ax2=sns.barplot(x=femaleRep.CONSTITUENCY,y=femaleRep.GENDER,ax=axes[1])
plt.sca(axes[1])
plt.ylabel('Female Candidate Count')

plt.show()
```

It can be observed that there is a huge disparity in the gender representation of candidates in different constituencies. The constituencies with highest number of male representation do not appear in the list for highest female representation at all.

#### Total Electors vs Votes

```python
# Checking for education-wise gender distribution
totalElectors = pd.DataFrame(lsElections.groupby(['CONSTITUENCY'])['TOTAL ELECTORS'].sum())
totalVotes = pd.DataFrame(lsElections.groupby(['CONSTITUENCY'])['TOTAL\nVOTES'].sum())
votePercentage = totalVotes.merge(totalElectors,on='CONSTITUENCY')
votePercentage['PERCENT VOTERS'] = votePercentage['TOTAL\nVOTES']/votePercentage['TOTAL ELECTORS']*100
votePercentage=votePercentage.sort_values(by=(['PERCENT VOTERS']),ascending=False).reset_index()
print(votePercentage.head())
```

```python
print(votePercentage.tail())
```

The highest voting percentage was seen in NOWGONG constituency with 40.75% eligible electors voting and least in ANANTNAG constituency with only 1.4% eligible electors voting.
